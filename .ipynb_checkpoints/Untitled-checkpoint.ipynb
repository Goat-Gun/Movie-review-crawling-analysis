```python
text=""
for i in tc['Î¶¨Î∑∞Î≥∏Î¨∏']:
    for j in i.split(" "):
        if ("series" in j) | ("Netflix" in j) | ("show" in j) | ("Squid" in j) | ("Game" in j) | ("episode" in j) | ("make" in j):
            pass
        else:
            text=text+" "+j
```
